```python
import unicodecsv as csv
import pandas as pd
```

```python
df = pd.DataFrame.from_csv("USCities.csv")

# create year column for grouping
df["year"] = df["dt"].apply(lambda x: x[:4])

df.drop_duplicates(subset=["Latitude", "Longitude", "year"], inplace=True)
del df["dt"]
del df["Country"]

df["AverageTemperature"] = pd.to_numeric(df["AverageTemperature"], errors="coerce")

df.groupby(["year", "City", "Latitude"])["AverageTemperature"].mean()
df.sort_values(by=["year", "City"], inplace=True)
df.to_csv("USCityAverageTempByYear.csv")
```
